### 2)

```python
ords_prods_merge.groupby('department_id').agg({'order_number' : ['mean','min','max']})
```

Overall, the results are not the same. Some departments had a lower average than from the subset, no more than 3 though. And other departments saw an increase in their mean. For example in department 10, it saw an increase some 18.6 to 20.2.

### 4)

```python
# Create a order for the max number of orders from customers
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Create a flag column to determine a customers loyalty 
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular Customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Check to see of column was created
ords_prods_merge[['max_order','loyalty_flag']].head()
```

```python
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
ords_prods_merge.groupby('loyalty_flag').agg({'prices':['min','max','mean']})
```

### 5)

```python
# Get basic stats on the prices of the various loyalty levels
ords_prods_merge.groupby('loyalty_flag').agg({'prices' : ['mean', 'min', 'max', 'median']})
```

As you can see there is a pricing difference between the different customers. It seems loyal customers tend to buy things at a lower price than other customers. This could be due to loyal customers looking for sales at the store and buying more items at a lower cost

### 6)

```python
# Create flag column of customers spending
ords_prods_merge['avg_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.loc[ords_prods_merge['avg_price'] < 10, 'spending_flag'] = 'Low spender'
ords_prods_merge.loc[ords_prods_merge['avg_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# Double to see if columns was created
ords_prods_merge.head()
```

```python
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

### 7)

```python
# Create flag column of customers order frequency
ords_prods_merge['median_prior_order'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.loc[ords_prods_merge['median_prior_order'] > 20, 'frequency_flag'] = 'Non-frequent customer'
ords_prods_merge.loc[(ords_prods_merge['median_prior_order'] <= 20) & (ords_prods_merge['median_prior_order'] > 10), 'frequency_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['median_prior_order'] <= 10, 'frequency_flag'] = 'Frequent Customer'
```

```python
# Check the frequency flag
ords_prods_merge.head()
```

```python
ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```
